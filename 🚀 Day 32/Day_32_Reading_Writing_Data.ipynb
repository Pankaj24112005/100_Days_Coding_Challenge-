# Day 32 – Reading & Writing Data (CSV, Excel, JSON)

## 1. Import Libraries

```python
import pandas as pd
```

## 2. Reading CSV Files

```python
df_csv1 = pd.read_csv('globalAirQuality.csv')
df_csv1
```

## 3. Handling Missing Values while Importing

```python
df_csv1.isna().sum()   #checking sum of null values in dataframe
```

## 4. Writing Data to CSV

```python
df_csv1.to_csv('globalAirQuality.csv', index=False)   #writing csv file
print('CSV file written successfully')
```

## 5. Reading JSON Files

```python
df_json1 = pd.read_json('employee_data.json')  #opening json file
df_json1.head()    #displaying first 5 rows data
```

## 6. Handling Missing Values while Importing

```python
df_json1.isna().sum()   #checking sum of null values in dataframe
```

## 7. Writing JSON Data

```python
df_json1.to_json('employee_data.json', orient='records', indent=2)
print('JSON file written successfully')
```

## 8. Reading Excel Files

```python
# Read CSV file
df = pd.read_csv("loan_prediction.csv")

# Save as Excel file
df.to_excel("loan_prediction.xlsx", index=False)

print("CSV converted to Excel successfully!")
df.head()   #displaying first 5 rows data
```

## 9. Handling Missing Values while Importing

```python
df.isna().sum()
```

## 8. When to use CSV vs Excel vs JSON
- **CSV**: Simple, fast, best for large tabular data
- **Excel**: Multiple sheets, business-friendly
- **JSON**: Nested data, APIs, web applications
