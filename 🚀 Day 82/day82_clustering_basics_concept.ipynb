# Day 82 – Clustering Basics

## What is Clustering?

Clustering is an **unsupervised machine learning technique** used to group similar data points together.

The goal is to place data points into clusters so that:
- Data points in the same cluster are similar
- Data points in different clusters are different

## Why Clustering is Important

Clustering helps in discovering hidden patterns in data.

It is widely used in:
- Customer segmentation
- Recommendation systems
- Image segmentation
- Anomaly detection

## Supervised vs Unsupervised Learning

| Type | Description |
|-----|-------------|
| Supervised Learning | Data has labels |
| Unsupervised Learning | Data has no labels |

Clustering belongs to **Unsupervised Learning**.

## Simple Example of Clustering

Imagine a dataset of customers.

Clustering can automatically group customers based on:
- Age
- Spending habits
- Purchase history

These groups help businesses understand different customer types.

## Visualization Example

Let's generate some random data to visualize how groups may appear.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

cluster1 = np.random.randn(50,2) + [2,2]
cluster2 = np.random.randn(50,2) + [7,7]
cluster3 = np.random.randn(50,2) + [12,2]

data = np.vstack((cluster1, cluster2, cluster3))

plt.scatter(data[:,0], data[:,1])
plt.title('Example Data That Could Form Clusters')
plt.show()
```

## Key Takeaways

- Clustering groups similar data points
- It is part of **unsupervised learning**
- Used to discover hidden patterns in datasets
