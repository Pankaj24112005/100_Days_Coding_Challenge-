# Day 69 – Linear Regression (Implementation from Scratch)

##  Import Libraries

```python

import numpy as np
import matplotlib.pyplot as plt
```

## Create Sample Dataset

```python

# Simple dataset (Hours studied vs Scores)
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

m = len(X)

print("X:", X)
print("y:", y)
```

## Initialize Parameters

```python

w = 0
b = 0

learning_rate = 0.01
iterations = 1000
```

## Define Cost Function (MSE)

```python

def compute_cost(X, y, w, b):
    total_cost = 0
    m = len(X)
    
    for i in range(m):
        prediction = w * X[i] + b
        total_cost += (prediction - y[i]) ** 2
        
    return total_cost / (2 * m)
```

## Implement Gradient Descent

```python

def gradient_descent(X, y, w, b, learning_rate, iterations):
    m = len(X)
    cost_history = []
    
    for _ in range(iterations):
        dw = 0
        db = 0
        
        for i in range(m):
            prediction = w * X[i] + b
            dw += (prediction - y[i]) * X[i]
            db += (prediction - y[i])
            
        dw = dw / m
        db = db / m
        
        w = w - learning_rate * dw
        b = b - learning_rate * db
        
        cost = compute_cost(X, y, w, b)
        cost_history.append(cost)
        
    return w, b, cost_history
```

## Train the Model

```python

w, b, cost_history = gradient_descent(X, y, w, b, learning_rate, iterations)

print("Final Weight (w):", w)
print("Final Bias (b):", b)
print("Final Cost:", cost_history[-1])
```

## Plot Regression Line

```python

predictions = w * X + b

plt.scatter(X, y)
plt.plot(X, predictions)
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.title("Linear Regression from Scratch")
plt.show()
```

## Plot Cost Reduction

```python

plt.plot(range(iterations), cost_history)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost Reduction over Iterations")
plt.show()
```
