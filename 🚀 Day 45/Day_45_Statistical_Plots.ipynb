```python

# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python

# Create a sample dataset
np.random.seed(42)

data = pd.DataFrame({
    "Math_Score": np.random.normal(70, 10, 200),
    "Science_Score": np.random.normal(65, 12, 200)
})

data.head()
```

## Box Plot

```python

plt.figure()
sns.boxplot(data=data)
plt.title("Box Plot of Student Scores")
plt.show()
```

##  Violin Plot

```python

plt.figure()
sns.violinplot(data=data)
plt.title("Violin Plot of Student Scores")
plt.show()
```

##  Distribution Plot

```python

plt.figure()
sns.histplot(data["Math_Score"], kde=True, label="Math", bins=30)
sns.histplot(data["Science_Score"], kde=True, label="Science", bins=30)
plt.legend()
plt.title("Distribution Plot of Scores")
plt.show()
```
