```python

import numpy as np
import pandas as pd
```

## Sample Dataset

```python

data = [10, 20, 20, 30, 40, 50, 50, 50, 60]

df = pd.DataFrame({"Values": data})
df
```

## Mean

```python

mean_value = df['Values'].mean()
mean_value
```

## Median

```python

median_value = df['Values'].median()
median_value
```

## Mode

```python

mode_value = df['Values'].mode()
mode_value
```

## Using NumPy

```python

np.mean(data), np.median(data)
```
