
# Day 72 – Logistic Regression (Theory)



## 1. What is Logistic Regression?

Logistic Regression is a **classification algorithm** used to predict the probability 
of a binary outcome (0 or 1).

Examples:
- Spam Detection
- Disease Prediction
- Loan Approval
- Customer Churn Prediction

Unlike Linear Regression, Logistic Regression predicts **probabilities**, not continuous values.



##  2. Sigmoid (Logistic) Function

Logistic Regression uses the sigmoid function to map any real-valued number 
into the range (0, 1).

$$
\sigma(z) = \frac{1}{1 + e^{-z}}
$$

Where:
- z = w₀ + w₁x₁ + w₂x₂ + ... + wₙxₙ

This output represents probability.



##  3. Why Not Use Linear Regression for Classification?

Linear Regression outputs values from -∞ to +∞.

Problems:
- Not bounded between 0 and 1
- Not suitable for probability interpretation
- Poor performance for classification tasks

Logistic Regression solves this by applying the sigmoid function.



## 4. Cost Function – Log Loss (Binary Cross-Entropy)

Instead of Mean Squared Error (MSE), Logistic Regression uses Log Loss:

$$
Loss = - \frac{1}{m} \sum [y \log(\hat{y}) + (1 - y)\log(1 - \hat{y})]
$$

Why?
- Ensures convex optimization
- Penalizes wrong confident predictions heavily



## 5. Decision Boundary

If:
- Probability ≥ 0.5 → Class 1
- Probability < 0.5 → Class 0

The model creates a **linear decision boundary** in feature space.



## 6. Assumptions of Logistic Regression

- Binary dependent variable
- Independent observations
- Little to no multicollinearity
- Large sample size preferred
- Linear relationship between independent variables and log-odds



## 7. Advantages & Limitations

### Advantages:
- Simple and interpretable
- Efficient to train
- Works well for linearly separable data

### Limitations:
- Not suitable for complex non-linear relationships
- Sensitive to outliers
