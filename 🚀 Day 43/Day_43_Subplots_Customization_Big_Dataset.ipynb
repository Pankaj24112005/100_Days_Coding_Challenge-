Matplotlib visualization with a large dataset.

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

np.random.seed(42)

data_size = 10000
data = {
    "Day": np.arange(1, data_size + 1),
    "Sales": np.random.randint(100, 1000, data_size),
    "Profit": np.random.randint(-200, 500, data_size),
    "Customers": np.random.randint(20, 200, data_size)
}

df = pd.DataFrame(data)
df.head()
```

```python

df.describe()
```

```python

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(df["Day"], df["Sales"])
plt.title("Sales Trend")
plt.xlabel("Day")
plt.ylabel("Sales")

plt.subplot(2, 2, 2)
plt.plot(df["Day"], df["Profit"])
plt.title("Profit Trend")
plt.xlabel("Day")
plt.ylabel("Profit")

plt.subplot(2, 2, 3)
plt.hist(df["Customers"], bins=30)
plt.title("Customer Distribution")
plt.xlabel("Customers")
plt.ylabel("Frequency")

plt.subplot(2, 2, 4)
plt.scatter(df["Sales"], df["Profit"])
plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")

plt.tight_layout()
plt.show()
```

```python

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

axes[0, 0].plot(df["Day"], df["Sales"])
axes[0, 0].set_title("Sales Over Time")

axes[0, 1].plot(df["Day"], df["Profit"])
axes[0, 1].set_title("Profit Over Time")

axes[1, 0].hist(df["Sales"], bins=40)
axes[1, 0].set_title("Sales Distribution")

axes[1, 1].scatter(df["Customers"], df["Profit"])
axes[1, 1].set_title("Customers vs Profit")

for ax in axes.flat:
    ax.set_xlabel("X-axis")
    ax.set_ylabel("Y-axis")
    ax.grid(True)

plt.tight_layout()
plt.show()
```
