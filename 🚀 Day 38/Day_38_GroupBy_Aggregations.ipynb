##  Import Required Libraries


```python

import pandas as pd
```


##  Create Sample Dataset


```python

data = {
    'Department': ['IT', 'IT', 'HR', 'HR', 'Finance', 'Finance'],
    'Employee': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Salary': [60000, 65000, 45000, 48000, 70000, 72000],
    'Experience': [2, 3, 1, 2, 4, 5]
}

df = pd.DataFrame(data)
df
```


## GroupBy Single Column


```python

df.groupby('Department')['Salary'].mean()
```


## Multiple Aggregations


```python

df.groupby('Department')['Salary'].agg(['min', 'max', 'mean', 'count'])
```


## GroupBy Multiple Columns


```python

df.groupby(['Department', 'Experience'])['Salary'].mean()
```


## Using agg() with Custom Names


```python

df.groupby('Department').agg(
    Avg_Salary=('Salary', 'mean'),
    Max_Salary=('Salary', 'max'),
    Total_Employees=('Employee', 'count')
)
```
