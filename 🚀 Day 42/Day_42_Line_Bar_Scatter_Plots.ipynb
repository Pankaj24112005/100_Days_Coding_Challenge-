# Day 42 – Line, Bar & Scatter Plots


```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Creating a dataset with 50 rows and 4 columns
np.random.seed(42)

data = {
    "Day": range(1, 51),
    "Sales": np.random.randint(100, 500, 50),
    "Profit": np.random.randint(20, 200, 50),
    "Customers": np.random.randint(10, 100, 50)
}

df = pd.DataFrame(data)
df.head()
```

## Line Plot – Sales Trend

```python

plt.figure()
plt.plot(df["Day"], df["Sales"])
plt.xlabel("Day")
plt.ylabel("Sales")
plt.title("Sales Trend Over Time")
plt.show()
```

## Bar Plot – Profit for First 10 Days

```python
# Bar Plot
plt.figure()
plt.bar(df["Day"][:10], df["Profit"][:10])
plt.xlabel("Day")
plt.ylabel("Profit")
plt.title("Profit for First 10 Days")
plt.show()
```

## Scatter Plot – Customers vs Sales

```python
# Scatter Plot
plt.figure()
plt.scatter(df["Customers"], df["Sales"])
plt.xlabel("Customers")
plt.ylabel("Sales")
plt.title("Customers vs Sales")
plt.show()
```
