# Pandas Series vs DataFrame

## Pandas Series – 

```python

import pandas as pd

# Example 1: Series from list
s1 = pd.Series([10, 20, 30, 40])
s1
```

```python

# Example 2: Series with custom index
s2 = pd.Series([100, 200, 300], index=['a', 'b', 'c'])
s2
```

```python

# Example 3: Series from dictionary
s3 = pd.Series({'Math': 85, 'Science': 90, 'English': 88})
s3
```

```python

# Example 4: Accessing elements
s1[0], s2['b']
```

## Pandas DataFrame – 

```python

# Example 1: DataFrame from dictionary
df1 = pd.DataFrame({
    'Name': ['Amit', 'Pankaj', 'Ravi'],
    'Age': [21, 22, 20]
})
df1
```

```python

# Example 2: DataFrame from list of lists
df2 = pd.DataFrame(
    [[1, 'Python'], [2, 'Java'], [3, 'C++']],
    columns=['ID', 'Language']
)
df2
```

```python

# Example 3: Accessing columns and rows
df1['Name'], df1.iloc[0]
```

```python

# Example 4: Adding a new column
df1['Marks'] = [85, 90, 88]
df1
```
