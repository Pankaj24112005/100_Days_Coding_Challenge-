```python

import pandas as pd
```

## Sample DataFrames

```python

df1 = pd.DataFrame({
    'id': [1, 2, 3],
    'name': ['A', 'B', 'C']
})

df2 = pd.DataFrame({
    'id': [2, 3, 4],
    'score': [80, 90, 85]
})

df1, df2
```

## Merge (Inner Join)

```python

pd.merge(df1, df2, on='id', how='inner')
```

## Merge (Left, Right, Outer)

```python

pd.merge(df1, df2, on='id', how='left')
```

```python

pd.merge(df1, df2, on='id', how='right')
```

```python

pd.merge(df1, df2, on='id', how='outer')
```

## Join using Index

```python

df1_indexed = df1.set_index('id')
df2_indexed = df2.set_index('id')

df1_indexed.join(df2_indexed)
```

## Concatenate (Vertical)

```python

pd.concat([df1, df1], axis=0)
```

## Concatenate (Horizontal)

```python

pd.concat([df1, df2], axis=1)
```
