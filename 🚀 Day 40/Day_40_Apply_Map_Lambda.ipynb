##  Lambda Functions
Lambda functions are small anonymous functions written in a single line.

```python
# Lambda function example
square = lambda x: x * x
square(5)
```

##  map() Function
The map() function applies a function to each element of an iterable.

```python
numbers = [1, 2, 3, 4, 5]

# Using map with lambda
squared_numbers = list(map(lambda x: x**2, numbers))
squared_numbers
```

##  apply() Function in Pandas
apply() is used to apply a function row-wise or column-wise in a DataFrame.

```python
import pandas as pd

df = pd.DataFrame({
    'A': [10, 20, 30],
    'B': [5, 15, 25]
})

df
```

```python
# Column-wise apply
df['Sum'] = df.apply(lambda row: row['A'] + row['B'], axis=1)
df
```

## apply() on Columns

```python
# Apply on columns
abc=df.apply(lambda col: col.max())
abc
```
