```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
print("Libraries Imported")
```

```python

# Load Dataset
df = pd.read_csv("train.csv")
df.head()
```

```python
df.columns
```

```python

df.info()
```

```python

df.describe()
```

## Numerical vs Numerical

```python
sns.scatterplot(data=df, x="Age", y="Fare")
plt.title("Age vs Fare")
plt.show()

df[["Age","Fare"]].corr()

```

## Categorical vs Numerical

```python
sns.boxplot(data=df, x="Sex", y="Fare")
plt.title("Sex vs Fare")
plt.show()


df.groupby("Sex")["Fare"].mean()
```

```python
# Example: Pclass vs Age
sns.boxplot(data=df, x="Pclass", y="Age")
plt.title("Pclass vs Age")
plt.show()

df.groupby("Pclass")["Age"].mean()
```

## Categorical vs Categorical

```python
# Example: Sex vs Survived

cross = pd.crosstab(df["Sex"], df["Survived"])
cross

sns.heatmap(cross, annot=True, fmt="d")
plt.title("Sex vs Survived")
plt.show()
```

## Pairplot

```python
sns.pairplot(df[["Age","Fare","SibSp","Parch","Survived"]])
plt.show()
```


## Observations
1. Fare and Age show weak correlation.
2. Females had higher survival rate than males.
3. Passengers in Pclass 1 survived more.
4. Higher Fare generally relates to higher survival.

### Conclusion:

Bivariate analysis reveals strong relationships between Sex, Pclass and Survived, which are important features for model building.
