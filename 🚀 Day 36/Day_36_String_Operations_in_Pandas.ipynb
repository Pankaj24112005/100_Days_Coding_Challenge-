# Day 36 â€“ String Operations in Pandas


```python
import pandas as pd

data = {
    'Names': [' Alice ', 'BOB', 'charlie', 'David '],
    'Emails': ['alice@gmail.com', 'bob@yahoo.com', 'charlie@gmail.com', 'david@outlook.com']
}

df = pd.DataFrame(data)
df
```

## Convert strings to lowercase and uppercase

```python
df['Names_lower'] = df['Names'].str.lower()
df['Names_upper'] = df['Names'].str.upper()
df
```

## Remove leading and trailing whitespaces

```python
df['Names_stripped'] = df['Names'].str.strip()
df
```

## Replace text

```python
df['Emails_replaced'] = df['Emails'].str.replace('gmail.com', 'company.com')
df
```

## Split strings

```python
df[['Username', 'Domain']] = df['Emails'].str.split('@', expand=True)
df
```

## Check if string contains a pattern

```python
df['Is_Gmail'] = df['Emails'].str.contains('gmail')
df
```
