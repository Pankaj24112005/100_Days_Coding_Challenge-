# Day 68 – Linear Regression (Theory)

## What is Linear Regression?
Linear Regression is a supervised learning algorithm used to predict a continuous target variable based on one or more input features.

It models the relationship between independent variable(s) (X) and dependent variable (Y).

## Simple Linear Regression Formula
y = mx + c

Where:
- m → Slope (coefficient)
- c → Intercept
- x → Input feature
- y → Predicted output

## Multiple Linear Regression Formula
y = b₀ + b₁x₁ + b₂x₂ + ... + bₙxₙ

Used when there are multiple input features.

## Cost Function
The most commonly used cost function in Linear Regression is **Mean Squared Error (MSE)**.

MSE = (1/n) Σ (Actual - Predicted)²

Goal: Minimize this error.

## Gradient Descent
Gradient Descent is an optimization algorithm used to minimize the cost function.

Steps:
1. Initialize parameters (m, c)
2. Compute error
3. Update parameters using learning rate
4. Repeat until convergence

## Evaluation Metrics
- R² Score
- Mean Absolute Error (MAE)
- Mean Squared Error (MSE)
- Root Mean Squared Error (RMSE)

## Assumptions of Linear Regression
1. Linearity
2. Independence of errors
3. Homoscedasticity
4. Normal distribution of residuals

## Conclusion
Linear Regression is one of the most fundamental algorithms in Machine Learning. Mastering it builds a strong foundation for advanced models like Decision Trees, Random Forests, and Neural Networks.
