# Day 62 – Variance & Standard Deviation

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Sample Dataset

```python

data = [10, 12, 14, 15, 18, 20, 22, 25]
df = pd.DataFrame({"Values": data})
df
```

## Mean, Variance & Standard Deviation

```python

mean = np.mean(data)
variance = np.var(data)
std_dev = np.std(data)

mean, variance, std_dev
```

## Using Pandas

```python

df.mean(), df.var(), df.std()
```

## Visualization

```python
plt.figure()
plt.hist(data, bins=6)

plt.title("Distribution of Data")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.grid(True)
plt.show()

```


### Conclusion
Variance and Standard Deviation help understand data spread and consistency — essential for machine learning and statistical analysis.
