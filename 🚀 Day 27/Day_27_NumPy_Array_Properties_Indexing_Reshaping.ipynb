## Array Properties, Indexing, Slicing, Reshaping & Manipulation

```
# Importing NumPy library
import numpy as np
```

## Creating NumPy Arrays

```
# 1D array
arr1 = np.array([10, 20, 30, 40, 50])

# 2D array
arr2 = np.array([[1, 2, 3],
                 [4, 5, 6]])

print(arr1)
print(arr2)
```

## Array Properties

```
print("Shape:", arr2.shape)
print("Dimensions:", arr2.ndim)
print("Size:", arr2.size)
print("Data Type:", arr2.dtype)
```

## Indexing

```
print(arr1[0])          # First element of 1D array
print(arr2[1, 2])       # Element from 2D array
print(arr2[0])          # First row
print(arr2[:, 1])       # Second column
```

## Slicing

```
print(arr1[1:4])        # Slice 1D array
print(arr2[0:2])        # Slice rows
print(arr2[:, 0:2])    # Slice columns
```

## Reshaping

```
reshaped_arr = arr1.reshape(5, 1)
print(reshaped_arr)

flat_arr = arr2.flatten()
print(flat_arr)

ravel_arr = arr2.ravel()
print(ravel_arr)
```

## Array Manipulation

```
print(arr2.T)            # Transpose
print(arr1 + 5)          # Addition
print(arr1 * 2)          # Multiplication
```

## Copy vs View

```
copy_arr = arr1.copy()
copy_arr[0] = 999
print("Original:", arr1)
print("Copy:", copy_arr)

view_arr = arr1.view()
view_arr[1] = 888
print("Original after view:", arr1)
print("View:", view_arr)
```

## Key Learning

NumPy arrays are faster and more memory-efficient than Python lists due to vectorized operations.
