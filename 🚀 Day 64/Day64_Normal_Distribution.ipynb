# Day 64 â€“ Normal Distribution

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Generate random data
data = np.random.normal(loc=50, scale=10, size=1000)
```

```python
# Plot histogram
plt.hist(data, bins=30, density=True)
```

```python
# Fit normal curve
mu, std = norm.fit(data)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p)
plt.title("Normal Distribution")
plt.xlabel("Values")
plt.ylabel("Density")
plt.show()
```

```python
print(f"Mean: {mu:.2f}")
print(f"Standard Deviation: {std:.2f}")
```

## Conclusion

- Data follows a bell-shaped curve.
- Mean and standard deviation describe distribution.
- Normal Distribution is essential in statistics and machine learning.
