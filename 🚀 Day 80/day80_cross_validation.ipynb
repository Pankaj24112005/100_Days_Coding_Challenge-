# Day 80 – Cross Validation

## What is Cross Validation?
Cross Validation is a technique used to evaluate a machine learning model by dividing the dataset into multiple parts and training/testing the model multiple times.

It helps produce a **more reliable estimate of model performance** compared to a single train-test split.

```python
# Import required libraries
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

## Load Dataset
We will use the Iris dataset for demonstration.

```python
# Load dataset
iris = load_iris()
X = iris.data
y = iris.target

print('Feature Shape:', X.shape)
print('Target Shape:', y.shape)
```

## Apply K-Fold Cross Validation

```python
# Initialize model
model = LogisticRegression(max_iter=200)

# Perform 5-fold cross validation
scores = cross_val_score(model, X, y, cv=5)

print('Cross Validation Scores:', scores)
print('Average Accuracy:', scores.mean())
```

## Conclusion
- Cross Validation improves model evaluation reliability.
- K-Fold Cross Validation trains and tests the model multiple times.
- The final performance is calculated as the **average of all scores**.
