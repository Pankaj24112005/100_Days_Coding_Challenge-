## Import NumPy

```python
import numpy as np
```

## NumPy Random Module
Generate random numbers, arrays, and values from distributions.

```python
np.random.seed(42)  # for reproducibility

# Random float between 0 and 1
rand_float = np.random.rand()
rand_float
```

```python
# Random integers
rand_ints = np.random.randint(1, 100, size=10)
rand_ints
```

```python
# Random array from normal distribution
rand_normal = np.random.randn(5)
rand_normal
```

## Statistical Functions in NumPy

```python
data = np.array([10, 20, 30, 40, 50])

mean = np.mean(data)
median = np.median(data)
std_dev = np.std(data)
variance = np.var(data)
min_val = np.min(data)
max_val = np.max(data)

mean, median, std_dev, variance, min_val, max_val
```
