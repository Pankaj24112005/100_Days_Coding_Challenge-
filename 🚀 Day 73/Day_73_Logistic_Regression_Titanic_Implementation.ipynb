
# Day 73 – Logistic Regression Implementation (Titanic Dataset)
Dataset: train_and_test2.csv


```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("train_and_test2.csv")

print("Shape:", data.shape)
data.head()
```

```python
# getting info about dataset
print(data.info())
print(data.describe())
print(data.isnull().sum())
```


## Data Cleaning

Observations:
- Many columns named 'zero', 'zero.1', etc. → redundant
- Target column name is '2urvived' → rename to 'Survived'
- Embarked has 2 missing values


```python
# Drop useless zero columns
zero_cols = [col for col in data.columns if "zero" in col]
data = data.drop(columns=zero_cols)

# Rename target column
data = data.rename(columns={"2urvived": "Survived"})

# Fill missing Embarked values
data["Embarked"] = data["Embarked"].fillna(data["Embarked"].mode()[0])

data.head()
```


## Feature Selection

Using important features:
- Pclass
- Sex
- Age
- Fare
- sibsp
- Parch
- Embarked


```python
features = ["Pclass", "Sex", "Age", "Fare", "sibsp", "Parch", "Embarked"]
X = data[features].values
y = data["Survived"].values

# Normalize features
X = (X - X.mean(axis=0)) / X.std(axis=0)

print("X shape:", X.shape)
print("y shape:", y.shape)
```


## Logistic Regression from Scratch


```python

class LogisticRegressionScratch:
    
    def __init__(self, lr=0.01, epochs=1000):
        self.lr = lr
        self.epochs = epochs
        
    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))
    
    def fit(self, X, y):
        m, n = X.shape
        self.weights = np.zeros(n)
        self.bias = 0
        
        for _ in range(self.epochs):
            linear_model = np.dot(X, self.weights) + self.bias
            y_pred = self.sigmoid(linear_model)
            
            dw = (1/m) * np.dot(X.T, (y_pred - y))
            db = (1/m) * np.sum(y_pred - y)
            
            self.weights -= self.lr * dw
            self.bias -= self.lr * db
    
    def predict(self, X):
        linear_model = np.dot(X, self.weights) + self.bias
        y_pred = self.sigmoid(linear_model)
        return np.array([1 if i > 0.5 else 0 for i in y_pred])
```

```python

model = LogisticRegressionScratch(lr=0.01, epochs=2000)
model.fit(X, y)

predictions = model.predict(X)

accuracy = np.mean(predictions == y)
print("Training Accuracy:", accuracy)
```
